```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# set the x-axis to a list of strings for each month
x_axis = ["Jan","Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()--this can allow you to change figure-level
#attributes or save figure as an image
#you are creating a reference by calling a variable named fig
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plot
#this is an example of using the object-oriented interface method.
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot for show
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label = 'Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set the y limit between 0 and 45
plt.ylim(0, 45)
plt.title("PyBer Fare by Month")
#add the legend
plt.legend()
```

```python
# Create the plot with markers and labels.
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth =2, label = 'Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set the y limit between 0 and 45
plt.ylim(0, 45)
plt.title("PyBer Fare by Month")
#add a grid
plt.grid()
#add the legend
plt.legend()
```

```python
#Create the plot with ax.plt()--this can allow you to change figure-level
#attributes or save figure as an image
#you are creating a reference by calling a variable named fig
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
y_axis_larger= []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#using the list comprehension technique, better for large datasets, it eliminates looping
plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
```

```python
# Using object oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Using object oriented interface skill drill
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=[i*5 for i in y_axis], edgecolors= 'b',label = 'Boston')
ax.legend()
```

```python
plt.pie(y_axis, labels = x_axis)
```

```python
explode_values =(0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize = (8,8))
plt.pie(y_axis, explode = explode_values, labels = x_axis, autopct = '%.1f%%')
```

```python
#Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green","yellowgreen","greenyellow", "yellow", "orange", "gold", "indianred", "tomato","mistyrose" ]
explode_values =(0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize = (8,8))
plt.pie(y_axis, 
        explode = explode_values,
        colors=colors,
        labels = x_axis, 
        autopct = '%.1f%%')
plt.show
```

```python
#skill drill for module 5.1.8
#object oriented approach
fig_dims = (8,8)
fig, ax = plt.subplots(figsize= fig_dims)
explode_values =(0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis,explode = explode_values, shadow = True, startangle = 90, counterclock = False, colors = colors,labels = x_axis, autopct = '%.1f%%')
```
